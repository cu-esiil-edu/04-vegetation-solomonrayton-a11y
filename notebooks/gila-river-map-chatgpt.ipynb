```python
import osmnx as ox
import geopandas as gpd
import hvplot.pandas
from shapely.geometry import Point

place = "Gila River Indian Community, Arizona, USA"

# Boundary
gila_boundary = ox.geocode_to_gdf(place)

# Roads
graph = ox.graph_from_place(
    place,
    network_type="drive"
)

roads = ox.graph_to_gdfs(
    graph,
    nodes=False
)


communities = gpd.GeoDataFrame(
    {
        "Name": [
            "Sacaton",
            "Casa Blanca",
            "Komatke"
        ]
    },
    geometry=[
        Point(-111.739, 33.076),
        Point(-111.889, 33.118),
        Point(-112.167, 33.294)
    ],
    crs="EPSG:4326"
)

# Layers
boundary_layer = gila_boundary.hvplot(
    geo=True,
    tiles="OSM",
    color=None,
    line_color="red",
    line_width=3,
    alpha=0.5
)

roads_layer = roads.hvplot(
    geo=True,
    color="gray",
    line_width=1
)

points_layer = communities.hvplot.points(
    geo=True,
    color="green",
    size=80,
    hover_cols=["Name"]
)

final_map = boundary_layer * roads_layer * points_layer

final_map
```
